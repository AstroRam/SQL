```python
from csvkit.utilities.csvsql import CSVSQL
```

```python
ChicagoCrimeData = CSVSQL('ChicagoCrimeData.csv')
ChicagoCensusData = CSVSQL('ChicagoCensusData.csv')
ChicagoPublicSchools = CSVSQL('ChicagoPublicSchools.csv')
```

```python
args_query1 = ['--query','select distinct(COUNT(ID)) FROM ChicagoCrimeData','ChicagoCrimeData.csv']

#Excute and print results of Query1
print("Result for the query:")

result1  = CSVSQL(args_query1)

print(result1.main())

```

```python
args_query1 = ['--query','SELECT COMMUNITY_AREA_NAME from ChicagoCensusData WHERE PER_CAPITA_INCOME < 11000;',
                  'ChicagoCensusData.csv']

#Excute and print results of Query1
print("Result for the query:")

result1  = CSVSQL(args_query1)

print(result1.main())

```

```python
# List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis)

args_query = ['--query','SELECT CASE_NUMBER FROM ChicagoCrimeData as C,
              ChicagoPublicSchools as P On P.COMMUNITY_AREA_NUMBER = C.COMMUNITY_AREA_NUMBER',
              'ChicagoPublicSchools.csv','ChicagoCrimeData.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
#List all kidnapping crimes involving a child?

args_query = ['--query','SELECT ID FROM ChicagoCrimeData WHERE PRIMARY_TYPE = KIDNAPPING ,
              'ChicagoCrimeData.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# What kinds of crimes were recorded at schools?

args_query = ['--query','SELECT PRIMARY_TYPE FROM ChicagoCrimeData as C,
              ChicagoPublicSchools as P WHERE C.COMMUNITY_AREA_NUMBER = P.COMMUNITY_AREA_NUMBER 
              GROUP By PRIMARY_TYPE ','ChicagoCrimeData.csv','ChicagoPublicSchools.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# List the average safety score for each type of school.

args_query = ['--query','SELECT Elementary FROM ChicagoPublicSchools GROUP BY Elementary',
              'ChicagoPublicSchools.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# List 5 community areas with highest % of households below poverty line 
# List the average safety score for each type of school.

args_query = ['--query','SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData 
              ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5',
              'ChicagoCensusData.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# Which community area is most crime prone?
args_query = ['--query','SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData 
              ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5','ChicagoCensusData.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# Use a sub-query to find the name of the community area with highest hardship index
args_query = ['--query','SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData 
              WHERE HARDSHIP_INDEX = (Select MAX(HARDSHIP_INDEX) 
              FROM ChicagoCensusData)','ChicagoCensusData.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```

```python
# Use a sub-query to determine the Community Area Name with most number of crimes?


args_query = ['--query','SELECT COMMUNITY_AREA_NAME,
              COUNT(1) as Num FROM ChicagoCrimeData as C, 
              ChicagoPublicSchools as P WHERE C.COMMUNITY_AREA_NUMBER = P.COMMUNITY_AREA_NUMBER
              GROUP By COMMUNITY_AREA_NAME ','ChicagoCrimeData.csv','ChicagoPublicSchools.csv']  

#Excute and print results of Query1
print("Result for the query:")

result  = CSVSQL(args_query)

print((result.main()))
```
